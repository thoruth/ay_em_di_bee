# Import stuff

```python
from magic_package import scraper, feature_helper
```

# download & scrape data

```python
df = scraper.get_data(verbose=1)
```

```python
df.columns
```

```python
rating_col = 'rating'
n_rating_col = 'n_rating'
n_oscar_col = 'n_oscar'
```

```python
df['n_oscar'] = df['n_oscar'].astype('int32')
```

# feature eng

```python
df['rating_calculated'] = df.apply(lambda x: feature_helper.rating_review_adjuster(x[rating_col], x[n_rating_col]), axis=1)
df['oscar_value'] = df.apply(lambda x: feature_helper.oscar_value(x[n_oscar_col]), axis=1)
```

```python
df['final_value']  = df['rating_calculated'] + df['oscar_value'] 
```
